# Model

```python
# we can apply linear regression model here
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
x= np.array(df['Hours']).reshape(-1,1)
y=np.array(df['Scores']).reshape(-1,1)
```

```python
model.fit(x,y)
```

```python
print(f'In this Linear Regression, the model score is :, {model.score(x,y):3.3f}')
```

Linear Regression Performance:  R2=0.953
